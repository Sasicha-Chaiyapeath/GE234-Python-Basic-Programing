## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile


```python
import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทยจากwebแม่โขง
gdf = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")

# แสดงข้อมูล 5 แถวแรก
print(gdf.head())
```


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python
import geopandas as gpd

# URL for province data (re-load to ensure gdf is fresh and has all columns)
url_province = "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
gdf = gpd.read_file(url_province)

# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร)
# หมายเหตุ: การคำนวณพื้นที่บน Geographic CRS (เช่น EPSG:4326) อาจไม่แม่นยำ
# ควร re-project ไปยัง Projected CRS ก่อนเพื่อความแม่นยำที่สูงขึ้น (เช่น UTM)

# Re-project to a suitable Projected CRS (e.g., UTM Zone 47N for Thailand, EPSG:32647)
gdf_projected = gdf.to_crs(epsg=32647)

# Calculate area in square meters and convert to square kilometers
gdf["area_sqkm"] = gdf_projected.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
print(gdf.nlargest(5, "area_sqkm")[["ADM1_TH", "area_sqkm"]])
```


## 🔹 ตัวอย่างที่ 5: ใช้ GeoPandas ทำ Spatial Join


```python

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```


# 📝 **กิจกรรมในแลป**

1. **แบบฝึกหัด 1**: ใช้ NumPy คำนวณค่า **Mean, Max, Min** ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง
2. **แบบฝึกหัด 2**: ใช้ GeoPandas โหลด **Shapefile** ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด
3. **แบบฝึกหัด 3**: ใช้ GeoPandas ทำ **Spatial Join** ระหว่างข้อมูลจังหวัดและอำเภอ
4. **แบบฝึกหัด 4**: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด


```python
#แบบฝึกหัด 1: ใช้ NumPy คำนวณค่า Mean, Max, Min ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

import numpy as np

# 1. สร้างข้อมูลตัวอย่าง (NIR และ Red)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# 2. คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

# 3. คำนวณค่าทางสถิติโดยใช้ NumPy Functions
mean_val = np.mean(ndvi)  # ค่าเฉลี่ย
max_val = np.max(ndvi)    # ค่าสูงสุด
min_val = np.min(ndvi)    # ค่าต่ำสุด

print("อาร์เรย์ NDVI:")
print(ndvi)
print("-" * 30)
print(f"ค่าเฉลี่ย (Mean): {mean_val:.2f}")
print(f"ค่าสูงสุด (Max): {max_val:.2f}")
print(f"ค่าต่ำสุด (Min): {min_val:.2f}")
```

```python
#แบบฝึกหัด 2: ใช้ GeoPandas โหลด Shapefile ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

import geopandas as gpd
import pandas as pd

# 1. โหลดข้อมูลจังหวัด
url = "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
gdf = gpd.read_file(url)

# 2. แปลงระบบพิกัดเป็น UTM Zone 47N (EPSG:32647) เพื่อให้หน่วยเป็นเมตร
# การคำนวณพื้นที่ในหน่วยองศา (4326) จะได้ค่าที่ไม่ถูกต้อง
gdf_projected = gdf.to_crs(epsg=32647)

# 3. คำนวณพื้นที่ (ตารางเมตร) แล้วแปลงเป็นตารางกิโลเมตร (หาร 1,000,000)
gdf["area_sqkm"] = gdf_projected.geometry.area / 1e6

# 4. ตั้งค่าให้ Pandas แสดงผลทุกแถว (กรณีต้องการดูครบ 77 จังหวัดในคอนโซล)
pd.set_option('display.max_rows', None)

# 5. เรียงลำดับจากพื้นที่มากไปน้อย และเลือกแสดงเฉพาะชื่อจังหวัดและพื้นที่
all_provinces = gdf[["ADM1_TH", "area_sqkm"]].sort_values(by="area_sqkm", ascending=False)

print("รายชื่อพื้นที่ทุกจังหวัดในประเทศไทย (ตร.กม.):")
print(all_provinces)

# หากต้องการบันทึกเป็นไฟล์ Excel หรือ CSV เพื่อดูข้อมูลทั้งหมด
# all_provinces.to_csv("thailand_provinces_area.csv", index=False)
```

```python
import geopandas as gpd

# 1. โหลดข้อมูลระดับจังหวัด (ADM1)
url_province = "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
gdf_provinces = gpd.read_file(url_province)

# 2. โหลดข้อมูลระดับอำเภอ (ADM2)
# แก้ไข URL ที่เสียไป โดยชี้ไปยังไฟล์ Shapefile อำเภอใน Google Drive ของคุณ
url_district = "/content/drive/MyDrive/Colab Notebooks/Amphoe/THA_Amphoe.shp"
gdf_districts = gpd.read_file(url_district)

# ตรวจสอบชื่อคอลัมน์ใน gdf_districts
print("Columns in gdf_districts:", gdf_districts.columns)

# 3. ตรวจสอบให้แน่ใจว่าระบบพิกัด (CRS) ตรงกัน
gdf_districts = gdf_districts.to_crs(gdf_provinces.crs)

# 4. ทำ Spatial Join
# ตรวจสอบว่า "อำเภอ" (gdf_districts) ไหน อยู่ "ภายใน" จังหวัด (gdf_provinces) ไหน
gdf_joined = gpd.sjoin(gdf_districts, gdf_provinces, how="inner", predicate="within")

# 5. แสดงผลลัพธ์ (คอลัมน์ A_NAME_T คือชื่ออำเภอ, ADM1_TH คือชื่อจังหวัด)
print(gdf_joined[['A_NAME_T', 'ADM1_TH', 'geometry']].head())
```

```python
#แบบฝึกหัด 4: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

import pandas as pd
import numpy as np
import geopandas as gpd

# 1. โหลดข้อมูล NDVI ของจริงที่คุณอัปโหลดมา
csv_file = "/content/drive/MyDrive/Colab Notebooks/tha-ndvi-subnat-full.csv"
df = pd.read_csv(csv_file)

# เลือกเฉพาะข้อมูลระดับจังหวัด (adm_level = 1)
df_prov = df[df['adm_level'] == 1].copy()

# 2. ใช้ NumPy (ผ่าน Pandas groupby) หาค่าเฉลี่ย NDVI (vim_avg) ของแต่ละจังหวัดตลอดกาล
# เราจัดกลุ่มตาม PCODE (รหัสจังหวัด เช่น TH95)
mean_ndvi = df_prov.groupby('PCODE')['vim_avg'].agg(np.mean).reset_index()

# สร้างคอลัมน์รหัสจังหวัดเป็นตัวเลข (ตัดคำว่า 'TH' ออกเพื่อให้เหลือเช่น '95')
# เพื่อเตรียมนำไป Join กับแผนที่
mean_ndvi['pro_code'] = mean_ndvi['PCODE'].str.replace('TH', '')

# 3. โหลดข้อมูลขอบเขตจังหวัด (GeoPandas)
url_geojson = "https://raw.githubusercontent.com/chingchai/OpenGISData-Thailand/master/provinces.geojson"
gdf_map = gpd.read_file(url_geojson)

# 4. ทำ Attribute Join เพื่อรวมข้อมูลตาราง NDVI เข้ากับแผนที่
# โดยเชื่อมผ่านรหัสจังหวัด (pro_code)
gdf_merged = gdf_map.merge(mean_ndvi, left_on='pro_code', right_on='pro_code', how='left')

# 5. หาจังหวัดที่มี NDVI เฉลี่ยสูงที่สุด
# ใช้ nlargest เพื่อดึงอันดับ 1
top_province = gdf_merged.nlargest(1, 'vim_avg')

print("--- รายงานข้อมูล NDVI (จากไฟล์จริง) ---")
print(f"จังหวัดที่มีค่า NDVI เฉลี่ยสูงสุดคือ: {top_province.iloc[0]['pro_th']}")
print(f"รหัสจังหวัด: {top_province.iloc[0]['PCODE']}")
print(f"ค่าความเขียว (NDVI) เฉลี่ย: {top_province.iloc[0]['vim_avg']:.4f}")

# 6. แสดง 5 อันดับจังหวัดที่เขียวที่สุดในประเทศไทย
print("\n--- 5 อันดับจังหวัดที่มีค่า NDVI สูงสุด ---")
top_5 = gdf_merged.nlargest(5, 'vim_avg')[['pro_th', 'PCODE', 'vim_avg']]
print(top_5.to_string(index=False))
```
