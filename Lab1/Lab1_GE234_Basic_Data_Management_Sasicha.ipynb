
### **4.2 การรวมข้อมูลเชิงพื้นที่และตาราง**
รวมข้อมูล Shapefile กับข้อมูลประชากรเพื่อแสดงข้อมูลประชากรในแต่ละจังหวัด โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

#1.จัดการข้อมูลประชากร (stat_c67.csv)
shapefile_path = '/content/drive/MyDrive/Colab Notebooks/tha_admbnda_adm1_rtsd_20190221/tha_admbnda_adm1_rtsd_20190221.shp'
gdf = gpd.read_file(shapefile_path)
# Corrected file path for df_stat
df_stat = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/stat_c67.csv', header=None, skiprows=1, dtype=str)

# ตั้งชื่อคอลัมน์ใหม่ (อ้างอิงจากการแกะข้อมูล)
df_stat = df_stat[[1, 11]].copy()
df_stat.columns = ['Province_Code', 'Population']

# แปลงข้อมูลให้เป็นตัวเลขและลบลูกน้ำ
def clean_currency(x):
    if isinstance(x, str):
        return x.replace(',', '').replace('"', '').strip()
    return x

df_stat['Population'] = df_stat['Population'].apply(clean_currency).astype(float)
df_stat['Population_Million'] = df_stat['Population'] / 1_000_000  # แปลงเป็นหน่วยล้านคน

#2.อ่านไฟล์ข้อมูลประชากร
file_path = '/content/drive/MyDrive/Colab Notebooks/stat_c67.csv'
df_pop = pd.read_csv(file_path, skiprows=4)

df_adm = gdf[['ADM1_PCODE', 'ADM1_EN']].copy() # Initialize df_adm from gdf

# ตัดคำว่า 'TH' ออกจากรหัสจังหวัดเพื่อให้ตรงกับไฟล์ stat (เช่น 'TH10' -> '10')
df_adm['Province_Code'] = df_adm['ADM1_PCODE'].str.replace('TH', '')

# ส่วนที่ 3: รวมข้อมูล (Merge Data)ษ
merged_df = pd.merge(df_adm, df_stat, on='Province_Code', how='inner')

# เปลี่ยนชื่อคอลัมน์ใน geo_data ให้ Merge ง่ายขึ้น
gdf = gdf.rename(columns={'ADM1_EN': 'ADM1_EN'}) # Ensure column name is consistent for merge

final_data = gdf.merge(merged_df, on='ADM1_EN', how='left') # Merge with gdf instead of geo_data

# ตรวจสอบจังหวัดที่ข้อมูลหาย (เป็น NaN)
missing = final_data[final_data['Population_Million'].isna()]['ADM1_EN']
if not missing.empty:
    print(f"จังหวัดที่ชื่อไม่ตรงกัน (ไม่มีข้อมูล): {missing.values}")

# พล็อตแผนที่
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

final_data.plot(
    column='Population_Million',
    cmap='YlOrRd',           # โทนสีเหลือง-ส้ม-แดง
    legend=True,
    legend_kwds={'label': "Population (Millions)", 'orientation': "horizontal", 'shrink': 0.5},
    missing_kwds={'color': 'lightgrey'},
    edgecolor='grey',
    linewidth=0.5,
    ax=ax
)

plt.title('Thailand Population Density 2024 (Official Data)', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()

# เครดิต
print('\n', "https://stat.bora.dopa.go.th/new_stat/webPage/statByYear.php")
print("https://data.opendevelopmentmekong.net/th/dataset/thailand-provincial-boundaries")
```

```python
#1. ลองเพิ่มประเทศใหม่ใน DataFrame พร้อมคำนวณความหนาแน่นประชากร (5 คะแนน)

import pandas as pd

# 1. โหลดไฟล์ CSV และกรองปี 2023
file_path = '/content/drive/MyDrive/Colab Notebooks/WB_WDI_SP_POP_TOTL.csv'
df = pd.read_csv(file_path)
df_csv = df
df_2023 = df_csv[df_csv['TIME_PERIOD'] == 2023]

# 2. เลือก 5 ประเทศนอกกลุ่ม ASEAN (ใช้รหัส ISO3)
selected_iso = ['FRA', 'JPN', 'BRA', 'USA', 'AUS'] # ฝรั่งเศส, ญี่ปุ่น, บราซิล, สหรัฐฯ, ออสเตรเลีย
df_task1 = df_2023[df_2023['REF_AREA'].isin(selected_iso)][['REF_AREA_LABEL', 'OBS_VALUE', 'REF_AREA']].copy()
df_task1.columns = ['Country', 'Population', 'ISO3']

# 3. เพิ่มข้อมูลพื้นที่ (Area) เนื่องจากในไฟล์ CSV ไม่มีข้อมูลนี้
areas = {
    'FRA': 551695, 'JPN': 377975, 'BRA': 8515767, 'USA': 9833520, 'AUS': 7692024
}
df_task1['Area_sq_km'] = df_task1['ISO3'].map(areas)

# 4. คำนวณความหนาแน่น
df_task1['Density'] = df_task1['Population'] / df_task1['Area_sq_km']
print(df_task1)

# เครดิต
print('\n',"Credit: https://data.worldbank.org/")
```

```python
# 2.นำไฟล์ Shapefile ของพื้นที่อื่นมาแสดงผลใน GeoPandas (5 คะแนน)
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

#1.อ่านไฟล์แผนที่ประเทศทั่วโลก (.shp)
shapefile_path = '/content/drive/MyDrive/Colab Notebooks/geoBoundariesCGAZ_ADM0/geoBoundariesCGAZ_ADM0.shp'
gdf = gpd.read_file(shapefile_path)

#2.อ่านไฟล์ข้อมูลประชากร
file_path = '/content/drive/MyDrive/Colab Notebooks/API_SP.POP.TOTL_DS2_en_csv_v2_40826/API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv'
df_pop = pd.read_csv(file_path, skiprows=4)

#3.รวมข้อมูลแผนที่กับข้อมูลประชากรโดยใช้ 'shapeName' และ 'Country Name'
# เนื่องจาก 'GID_0' ไม่พบใน gdf
merged_gdf = gdf.merge(df_pop, left_on='shapeName', right_on='Country Name', how='left')

# ตรวจสอบผลลัพธ์ของการรวมข้อมูลเบื้องต้น
print(merged_gdf.head())
# 4. กำหนดปีที่ต้องการแสดงข้อมูล (ต้องตรงกับชื่อคอลัมน์ในไฟล์ CSV เช่น '2020', '2021', '2022')
year_col = '2022'

# ตรวจสอบว่ามีข้อมูลปีนั้นจริงหรือไม่ เพื่อกัน Error
if year_col in merged_gdf.columns:

# 5. เริ่มวาดแผนที่
    fig, ax = plt.subplots(1, 1, figsize=(15, 10)) # กำหนดขนาดรูป

    merged_gdf.plot(
        column=year_col,           # คอลัมน์ที่จะใช้แบ่งเฉดสี (จำนวนประชากร)
        ax=ax,                     # วาดลงบนแกนที่เตรียมไว้
        legend=True,               # แสดงแถบสีบอกค่า
        legend_kwds={'label': "Population Count", 'orientation': "horizontal"},
        cmap='Spectral_r',         # โทนสี (ลองเปลี่ยนเป็น 'OrRd', 'viridis' ได้)
        missing_kwds={
            "color": "lightgrey",  # สีของประเทศที่ไม่มีข้อมูล (Merge ไม่ติด)
            "label": "Missing values"
        }
    )

    # ตกแต่งชื่อกราฟและปิดแกน
    ax.set_title(f'World Population Map ({year_col})', fontsize=16)
    ax.set_axis_off()

    # สำคัญ: คำสั่งให้แสดงผล
    plt.show()

# เครดิต
print('\n',"Credit: https://data.worldbank.org/")
print("Credit: https://www.geoboundaries.org/")
```

```python
# 3.ลองสร้างแผนที่ด้วยสีที่แตกต่างกันโดยใช้ค่าอื่น เช่น area หรือ density (5 คะแนน)
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

#1.อ่านไฟล์แผนที่ประเทศทั่วโลก (.shp)
shapefile_path = '/content/drive/MyDrive/Colab Notebooks/geoBoundariesCGAZ_ADM0/geoBoundariesCGAZ_ADM0.shp'
gdf = gpd.read_file(shapefile_path)

#2.อ่านไฟล์ข้อมูลประชากร
file_path = '/content/drive/MyDrive/Colab Notebooks/API_SP.POP.TOTL_DS2_en_csv_v2_40826/API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv'
df_pop = pd.read_csv(file_path, skiprows=4)

#3.รวมข้อมูลแผนที่กับข้อมูลประชากรโดยใช้ 'shapeName' และ 'Country Name'
# เนื่องจาก 'GID_0' ไม่พบใน gdf
merged_gdf = gdf.merge(df_pop, left_on='shapeName', right_on='Country Name', how='left')

# ตรวจสอบผลลัพธ์ของการรวมข้อมูลเบื้องต้น
print(merged_gdf.head())

#4.สร้างแผนที่ระบายสีตามค่าที่แตกต่างกัน (Density vs Area)
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

#5.ทดลองระบายสีตามประชากรปี 2023 (เนื่องจากคอลัมน์ Density (per km²) ไม่มีในข้อมูลนี้)
population_column = '2023'
if population_column in merged_gdf.columns:
    merged_gdf.plot(column=population_column,
                cmap='OrRd',          # ใช้โทนสีส้ม-แดง (OrRd)
                legend=True,
                legend_kwds={'label': "Population (2023)", 'orientation': "horizontal"},
                ax=ax)
    plt.title('World Population Map (2023)', fontsize=16)
    plt.axis('off') # ปิดแกนพิกัด
else:
    plt.title(f'World Population Map ({population_column} Data Missing)', fontsize=16)
    plt.axis('off')

plt.show()

# เครดิต
print('\n',"Credit: https://data.worldbank.org/")
print("Credit: https://www.geoboundaries.org/")
```
